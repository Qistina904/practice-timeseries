```python
#Feature Engineering 1
indiaclimate.head(7)
```

```python
myfeatures = indiaclimate.copy()
```

```python
myfeatures['year']= indiaclimate['date'].dt.year
myfeatures['month'] =indiaclimate['date'].dt.month
myfeatures['day'] = indiaclimate['date'].dt.day
```

```python
myfeatures.head()
```

```python
#determine lag features. like K step ft + K, K = lag 2 
myfeatures['lag2'] = indiaclimate['wind_speed'].shift(2)
```

```python
#1 year lag
myfeatures['lagyear'] = indiaclimate['wind_speed'].shift(365)
myfeatures.tail()
```

```python
#calc mean with window size of 2
myfeatures['mean_roll']=indiaclimate['wind_speed'].rolling(window=2).mean()
myfeatures.head()
```

```python
myfeatures['max_roll']=indiaclimate['wind_speed'].rolling(window=3).max()
myfeatures.head(10)
```

```python
#fo entire column, max value 
myfeatures['max_expand'] =indiaclimate['wind_speed'].expanding().max()
myfeatures.head(10)
```
